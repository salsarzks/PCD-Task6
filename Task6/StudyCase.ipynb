```python
#1. hitung dan menampilkan histogram dari citra asli.
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Baca gambar grayscale
img = cv2.imread('wonban.jpeg', 0)

# Histogram Equalization
equ = cv2.equalizeHist(img)

# Hitung histogram gambar asli
hist_img = cv2.calcHist([img], [0], None, [256], [0, 256])

# Tampilkan gambar asli
plt.figure(figsize=(15, 12))

# Gambar Asli
plt.subplot(4, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('Gambar Asli')
plt.axis('off')

plt.subplot(4, 2, 2)
plt.plot(hist_img, color='black')
plt.title('Histogram Gambar Asli')

plt.tight_layout()
plt.show()
```

```python
#2. lakukan histogram equalization pada citra.

import cv2
from matplotlib import pyplot as plt

# Baca gambar grayscale
img = cv2.imread('wonban.jpeg', 0)

# Histogram Equalization
equ = cv2.equalizeHist(img)

# Tampilkan gambar asli dan hasil equalization
plt.subplot(121), plt.imshow(img, cmap='gray'), plt.title('Gambar Asli')
plt.subplot(122), plt.imshow(equ, cmap='gray'), plt.title('Hasil Equalization')
plt.show()
```

```python
# 3.  hitung dan tampilkan histogram dari citra hasil equalization.
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Baca gambar grayscale
img = cv2.imread('wonban.jpeg', 0)

# Histogram Equalization
equ = cv2.equalizeHist(img)

# Hitung histogram hasil equalization
hist_equ = cv2.calcHist([equ], [0], None, [256], [0, 256])

# Tampilkan gambar equalization
plt.figure(figsize=(15, 12))

# Hasil Equalization
plt.subplot(4, 2, 3)
plt.imshow(equ, cmap='gray')
plt.title('Hasil Equalization')
plt.axis('off')

plt.subplot(4, 2, 4)
plt.plot(hist_equ, color='black')
plt.title('Histogram Hasil Equalization')
plt.tight_layout()
plt.show()
```

```python
#4. bandingkan citra asli dan citra hasil equalization untuk melihat perbaikan kontras
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Baca gambar grayscale
img = cv2.imread('wonban.jpeg', 0)

# Histogram Equalization
equ = cv2.equalizeHist(img)

# Hitung histogram gambar asli, hasil equalization
hist_img = cv2.calcHist([img], [0], None, [256], [0, 256])
hist_equ = cv2.calcHist([equ], [0], None, [256], [0, 256])

# Tampilkan gambar asli, equalization dan histogram masing-masing
plt.figure(figsize=(15, 12))

# Gambar Asli
plt.subplot(4, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('Gambar Asli')
plt.axis('off')

plt.subplot(4, 2, 2)
plt.plot(hist_img, color='black')
plt.title('Histogram Gambar Asli')

# Hasil Equalization
plt.subplot(4, 2, 3)
plt.imshow(equ, cmap='gray')
plt.title('Hasil Equalization')
plt.axis('off')

plt.subplot(4, 2, 4)
plt.plot(hist_equ, color='black')
plt.title('Histogram Hasil Equalization')
plt.tight_layout()
plt.show()
```
